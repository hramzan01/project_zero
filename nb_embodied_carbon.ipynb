### Loading the datasource

```python
import pandas as pd
```

```python
df_model = pd.read_csv('data_city_hz.csv')

#read the unique asset types
df_model['RSHP_AD_IdAsset'].unique()
```

```python
# custom function to assign structure per type
def facade_type(row):
    if row['RSHP_AD_IdAsset'] == 'HOTEL':
        return 'aluminium/steel'
    return 'undefined'

# apply function to each rown in df
df_model['Envelope Material'] = df_model.apply(facade_type, axis=1)    

#display the dataframe
df_model.head()
```
